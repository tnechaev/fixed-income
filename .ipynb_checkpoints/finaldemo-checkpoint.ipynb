```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from fixed_income_toolkit import FixedIncomeToolkit
```

```python
toolkit = FixedIncomeToolkit()
# init fetcher if you want live ECB data 
toolkit.init_ecb_fetcher()

maturities_cols = [
    "SR_3M","SR_6M","SR_1Y","SR_2Y","SR_3Y",
    "SR_5Y","SR_7Y","SR_10Y","SR_15Y","SR_20Y","SR_30Y"
]

df_yields = toolkit.fetch_ecb_yield_curve(maturities=maturities_cols,
                                          start_period="2018-01-01", end_period="2023-12-31")

# Keep subset with no missing rows
df = df_yields[maturities_cols].dropna().copy() / 100.0
# convert column labels to numeric years
def tenor_to_years(x):
    x = x.replace("SR_","")
    if "M" in x: return float(x.replace("M",""))/12.0
    if "Y" in x: return float(x.replace("Y",""))
    return float(x)
maturities = np.array([tenor_to_years(c) for c in df.columns])
df.columns = maturities
```

```python
# 1) Fit NS per-date time series (OLS)
ns_ts = toolkit.fit_ns_time_series(df, maturities)
ns_ts[['beta0','beta1','beta2']].plot(subplots=True, figsize=(10,6), title="Nelson-Siegel betas (per-date fit)")
plt.show()

# 2) PCA on yield changes
pca = toolkit.pca_yield_changes(df)
print("Explained variances (first 3):", pca["explained"])
for i in range(min(3, pca["components"].shape[0])):
    plt.plot(maturities, pca["components"][i,:], marker='o', label=f"PC{i+1}")
plt.title("PCA loadings (yield changes)"); plt.xlabel("Maturity (yrs)"); plt.legend(); plt.show()
```

```python
def tune_ekf_qr(toolkit, yield_df, maturities,
                Q_grid, R_grid,
                train_size=500, test_horizon=50):
    """
    Rolling OOS tuning for EKF.
    Returns best (Q,R) and score table.
    """
    Y = yield_df.values
    T = len(Y)
    results = []

    for Q in Q_grid:
        for R in R_grid:
            errs = []
            for t0 in range(train_size, train_size + test_horizon):
                train = yield_df.iloc[:t0]
                test = yield_df.iloc[t0]

                smoothed, _ = toolkit.ekf_rts_smoother(
                    train,
                    Q_scale=Q,
                    R_scale=R,
                    init_tau=2.5
                )

                # one-step state forecast (random walk)
                x_last = smoothed[-1]
                y_hat = toolkit.ns_yields_from_state(x_last, maturities)

                rmse = np.sqrt(np.mean((y_hat - test.values)**2))
                errs.append(rmse)

            mean_rmse = np.mean(errs)
            results.append((Q, R, mean_rmse))
            print(f"Q={Q:.1e}, R={R:.1e} -> RMSE={mean_rmse:.6f}")

    res_df = pd.DataFrame(results, columns=["Q_scale","R_scale","RMSE"])
    best = res_df.loc[res_df["RMSE"].idxmin()]
    return best, res_df
```

```python
Q_grid = [1e-7, 5e-7, 1e-6, 5e-6, 1e-5]
R_grid = [1e-7, 5e-7, 1e-6, 5e-6, 1e-5]

best, table = tune_ekf_qr(
    toolkit,
    yield_df=df,
    maturities=maturities,
    Q_grid=Q_grid,
    R_grid=R_grid,
    train_size=500,
    test_horizon=50
)

print("\nBest parameters:")
print(best)
```

```python
def price_bonds_under_curves(toolkit, curves, maturities):
    prices = np.zeros((curves.shape[0], len(toolkit.bonds)))

    for i, curve in enumerate(curves):
        zero = lambda t: np.interp(t, maturities, curve)
        for j,b in enumerate(toolkit.bonds):
            p, *_ = toolkit.price_fixed_rate_bond(
                b["face"], b["coupon_rate"], b["maturity"], zero
            )
            prices[i,j] = p
    return prices

prices_0 = price_bonds_under_curves(
    toolkit,
    np.repeat(df.values[-1][None,:], future_curves.shape[0], axis=0),
    maturities
)

prices_1 = price_bonds_under_curves(
    toolkit, future_curves, maturities
)

returns = (prices_1 - prices_0) / prices_0
toolkit.scenario_returns = returns
```

```python
opt, res, extras = toolkit.minimize_cvar(alpha=0.95)

print("\nOptimal weights:")
for w,b in zip(opt["weights"], toolkit.bonds):
    print(f"{b['name']}: {w:.3f}")

print("\nPortfolio stats:")
print("Mean return:", opt["mean_return"])
print("CVaR(95%):", opt["cvar"])

##### BROKEN NEED TO FIX
```
